Lets load the dataset.

```python
# Loading Dataset
import kagglehub

# Download latest version
path = kagglehub.dataset_download("paperxd/cleaned-life-expectancy-dataset")

print("Path to dataset files:", path)
print("File name: Cleaned-Life-Exp.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = path + '/Cleaned-Life-Exp.csv'
filename
data = pd.read_csv(filename)
```

The dataset is already cleaned and preprocessed. so lets reformat our target variable. we want to convert life expectancy to categorical variable. we use bins to achieve this goal, exactly like what we did in last project (SVM algorithm).

```python
# Binning countinues life expectancy
bins = [-4,-1.5,0.5,3]
labels = ['Low','Medium','High']
data['lifeexp_category'] = pd.cut(data['Life expectancy'], bins = bins, labels = labels)
```

Lets create our features, target datasets. we also need to do one-hot encoding for Country column.

```python
X = data.drop(['Life expectancy','lifeexp_category'],axis=1)
X = pd.get_dummies(X, columns=['Country'], drop_first=True)
Y = data['lifeexp_category']
```

now we use sklearn train-test split for creating train and test data. we use test_size = 0.2 and random_state=2. first we need to import the package.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

Now its time for feature scaling.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Now we are ready to train the model. We also want to find optimal hyper-parameters using grid search. so first we create a dictionary to store all hyper-parameters and then we train model with different hyper-parameters to find the optimal ones.

```python
from sklearn.tree import DecisionTreeClassifier

dt_param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 5, 10, 25, 50],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 5, 7]
}
dt_model = DecisionTreeClassifier(random_state=42)
grid_search_dt = GridSearchCV(estimator=dt_model, param_grid=dt_param_grid, cv=5, verbose=2, n_jobs=-1)
grid_search_dt.fit(X_train_scaled, y_train)

# Best Decision Tree model
best_dt_model = grid_search_dt.best_estimator_
y_pred_dt = best_dt_model.predict(X_test_scaled)
```

Our optimal model is trained. lets print optimal hyper-parameters.

```python
print(f"Best parameters: {grid_search_dt.best_params_}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred_dt))

plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred_dt), annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

Thats it! now we want to run random forest with hyper-parameter tuning. 

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest Model and Hyperparameter Tuning
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10, 15, 17],
    'min_samples_leaf': [1, 2, 5, 7, 9]
}
rf_model = RandomForestClassifier(random_state=2)
grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=rf_param_grid, cv=5, verbose=2, n_jobs=-1)
grid_search_rf.fit(X_train_scaled, y_train)

# Best Random Forest model
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test_scaled)
```

```python
print("Best parameters for Random Forest:", grid_search_rf.best_params_)
```

```python
# Evaluation for Random Forest
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))

# Confusion Matrix for Random Forest
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d', cmap='Greens', xticklabels=labels, yticklabels=labels)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
